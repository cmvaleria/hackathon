```python
import pandas as pd
import numpy as np
```

```python
data_train = pd.read_csv('train.csv')
data_test = pd.read_csv('test_no_class.csv')
```

```python
cols = []
for col in data_train.columns:
  cols.append(col.replace(' ', '_'))
data_train.columns = cols
```

```python
cols = []
for col in data_test.columns:
  cols.append(col.replace(' ', '_'))
data_test.columns = cols
```

# Cleaning

### SEX

```python
data_train['SEX'] = data_train['SEX'].str.lower()
data_test['SEX'] = data_test['SEX'].str.lower()
```

```python
def get_gender_short(gender):
    if gender.lower().startswith('m'):
        return 'm'
    elif gender.lower().startswith('f'):
        return 'f'
    else:
        return None
    
data_train['SEX'] = data_train['SEX'].apply(get_gender_short)
data_test['SEX'] = data_test['SEX'].apply(get_gender_short)
```

```python
display(data_train.SEX.value_counts())
display(data_test.SEX.value_counts())
```

### K-NN imputing 

```python
data_train.replace('?', np.nan, inplace=True)
```

```python
data_test.replace('?', np.nan, inplace=True)
```

```python
data_train.isnull().sum()/len(data_train)
```

```python
data_test.isnull().sum()/len(data_train)
```

```python
from sklearn.impute import KNNImputer

df = data_train.drop(columns=['SEX','Class'])

imputer = KNNImputer(n_neighbors=2)

df_imputed_train = imputer.fit_transform(df)

df_imputed_train = pd.DataFrame(df_imputed_train, columns=df.columns)

df_imputed_train
```

```python
data_train = pd.concat([data_train[['AGE','SEX']], df_imputed_train, data_train[['Class']]], axis=1)
```

```python
data_train.isnull().sum()/len(data_train)
```

```python
from sklearn.impute import KNNImputer

df = data_test.drop(columns=['SEX'])

imputer = KNNImputer(n_neighbors=2)

df_imputed = imputer.fit_transform(df)

df_imputed = pd.DataFrame(df_imputed, columns=df.columns)
```

```python
data_test = pd.concat([data_test[['AGE','SEX']], df_imputed], axis=1)
```

```python
data_test.isnull().sum()/len(data_test)
```

```python
def clean_sex(x):
    if x in ['m']:
        return '1'
    else:
        return '0'
data_train['SEX'] = data_train['SEX'].apply(clean_sex)
data_test['SEX'] = data_test['SEX'].apply(clean_sex)
```

```python
display(data_train.info())
display(data_test.info())
```

```python
data_train['SEX'] = data_train['SEX'].astype(float)
data_test['SEX'] = data_test['SEX'].astype(float)
```

# Random Forest

#### TRAINING THE MODEL

```python
data_train = data_train.drop(columns=['ALK_PHOSPHATE','PROTIME'])
data_test = data_test.drop(columns=['ALK_PHOSPHATE','PROTIME'])
```

```python
X_train = data_train.drop('Class',axis='columns')
y_train = data_train.Class
```

```python
X_train = X_train.iloc[:,1:]
```

```python
X_test = data_test.iloc[:,1:]
```

```python
from imblearn.over_sampling import RandomOverSampler

ros = RandomOverSampler(sampling_strategy="not majority") # String
X_train_ros, y_train_ros = ros.fit_resample(X_train, y_train)
```

```python
y_train.value_counts()
```

```python
y_train_ros.value_counts()
```

```python
from sklearn.ensemble import RandomForestClassifier
model_v1 = RandomForestClassifier(n_estimators=20)
model_v1.fit(X_train_ros, y_train_ros)
```

```python
model_v1.score(X_train_ros, y_train_ros)
```

<font color ='red'> overfitting </font>

### Parameters:

```python

n_estimators = [int(x) for x in np.linspace(start = 10, stop = 80, num = 10)]

max_features = ['auto', 'sqrt']

max_depth = [2,4]

min_samples_split = [2, 5]

min_samples_leaf = [1, 2]

bootstrap = [True, False]
```

```python
param_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
print(param_grid)
```

```python
from sklearn.model_selection import GridSearchCV
rf_Grid = GridSearchCV(estimator = model_v1, param_grid = param_grid, cv = 3, verbose=2, n_jobs = 4)
```

```python
rf_Grid.fit(X_train_ros, y_train_ros)
```

```python
rf_Grid.best_params_
```

```python
print (f'Train Accuracy - : {rf_Grid.score(X_train_ros,y_train_ros):.3f}')
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 3)
knn.fit(X_train_ros, y_train_ros)
y_pred_train = knn.predict(X_train_ros)
```
